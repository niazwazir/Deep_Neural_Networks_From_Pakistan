```
import keras
from keras.models import Model
from keras.layers import Input, Dense
```

```
input = Input((None, 1))
output = Dense(1,use_bias=False)(input)
model = Model(input, output)
model.summary()
```

```
len(model.layers)
```

```
keras.utils.plot_model(model, 'my_first_model.png')
```

```
input = Input((None, 4))
dense = Dense(4,use_bias=False)(input)
dense1 = Dense(3,use_bias=False)(dense)
output = Dense(1,use_bias=False)(dense1)
model = Model(input, output)
model.summary()
```
